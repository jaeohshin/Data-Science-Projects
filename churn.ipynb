#Feature engineering


```python
#Convert Churn column to integer
df.loc[df.Churn == 'No', 'Churn'] = 0
df.loc[df.Churn == 'Yes', 'Churn'] = 1
```

```python
df["Churn"] = df.Churn.astype(float)
df.head()
df.info()
```

```python
df.groupby('gender').Churn.mean()
```

The churn rate of female is slightly (0.8%) higher than male's

```python
df_plot = df.groupby('gender').Churn.mean().reset_index()
plot_data = [
    go.Bar(x = df_plot['gender'],
           y = df_plot['Churn'],
           width = [0.5, 0.5],
           marker = dict(color=['green', 'blue'])
           )
]

plot_layout = go.Layout(
    xaxis = {"type": "category"},
    yaxis = {"title": "Churn Rate"},
    title = 'Gender'
)

fig = go.Figure(data = plot_data, layout = plot_layout)
pyoff.iplot(fig)
```

```python
df_plot = df.groupby('Partner').Churn.mean().reset_index()
plot_data = [
    go.Bar(x = df_plot['Partner'],
           y = df_plot['Churn'],
           width = [0.5, 0.5],
           marker = dict(color=['green', 'blue'])
           )
]

plot_layout = go.Layout(
    xaxis = {"type": "category"},
    yaxis = {"title": "Churn Rate"},
    title = 'Partner'
)

fig = go.Figure(data = plot_data, layout = plot_layout)
pyoff.iplot(fig)
```
